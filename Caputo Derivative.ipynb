```python
import numfraccal as nfr
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rc

rc('font',**{'family':'serif','serif':['Times New Roman']})
rc('text', usetex=True)
%matplotlib inline
```

# Caputo Derivative

nfr.fd_caputo(g,t,a,alpha,h)

Returns:

$\bullet$ Numerical Value of Caputo Derivative

$\bullet$ Numerical Error in integration

Notes:
h:differential in numerical evaluation of first derivative

<font color=red>Left</font> Caputo derivatives with order $0<\alpha<1$ of the given function $g(t)$ [see definition in Eq. (1.7) in Numerical Methods for Fractional Calculus]

$${}_C D_{a,t}^\alpha g(t)=\frac{1}{\Gamma(1-\alpha)}\int_a^t \frac{g^\prime(s)}{(t-s)^\alpha}ds$$

where $\Gamma(...)$ donotes the Gamma funtion and $g^\prime(t)$ is the first derivative of $g(t)$.

### Example: Fractional Derivative of $g(t)=t$

```python
def g(t):
    return t
```

```python
fig = plt.figure(figsize=(8,5))
plt.title(r'$g(t)=t$',y=1.01,size=20)
for alpha in np.linspace(0.0,0.9,6):
    T=[]
    Df=[]
    t=0.01
    while t<=10:
        T.append(t)
        Df.append(nfr.fd_Caputo(g,t,0,alpha,1e-6)[0])
        t+=0.01
    plt.plot(T,Df,label=r'$\alpha='+str(round(alpha,1))+'$',linewidth=2)
plt.legend(loc=(1,0.2),shadow=False,ncol=1,prop={'size':16},frameon=False)
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel(r'$t$',size=23)
plt.ylabel(r'${}_C D_{0,t}^\alpha\, g(t)$',size=23)
plt.show()
fig.savefig('Caputo1.png',dpi=100,bbox_inches='tight')
```

### Example: Fractional Derivative of $g(t)=t\sin(t)$

```python
def g(t):
    return t*np.sin(t)
```

```python
fig = plt.figure(figsize=(8,5))
plt.title(r'$g(t)=t\sin(t)$',y=1.01,size=20)
for alpha in np.linspace(0.1,0.9,9):
    T=[]
    Df=[]
    t=0.01
    while t<10:
        T.append(t)
        Df.append(nfr.fd_Caputo(g,t,0,alpha,1e-6)[0])
        t+=0.01
    plt.plot(T,Df,label=r'$\alpha='+str(round(alpha,1))+'$',linewidth=0.5+alpha*2)
plt.legend(loc=(1,0.2),shadow=False,ncol=1,prop={'size':16},frameon=False)
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel(r'$t$',size=23)
plt.ylabel(r'${}_C D_{0,t}^\alpha\, g(t)$',size=23)
plt.show()
fig.savefig('Caputo2.png',dpi=100,bbox_inches='tight')
```
