```python
import numfraccal as nfr
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rc

rc('font',**{'family':'serif','serif':['Times New Roman']})
rc('text', usetex=True)
%matplotlib inline
```

# Riemannâ€“Liouville Derivative

nfr.fd_RiemannLiouville(g,t,a,alpha,h)

Returns:

$\bullet$ Numerical Value of Riemann Liouville Derivative


Notes:
h:differential in numerical evaluation of first derivative

<font color=red>Left</font> Riemann Liouville derivatives with order $0<\alpha<1$ of the given function $g(t)$ [see definition in Eq. (1.5) in Numerical Methods for Fractional Calculus]

$${}_{RL} D_{a,t}^\alpha g(t)=\frac{1}{\Gamma(1-\alpha)}\frac{d}{dt}\int_a^t \frac{g(s)}{(t-s)^\alpha}ds$$

where $\Gamma(...)$ donetes the Gamma funtion.

### Example: Fractional Derivative of $g(t)=e^{-t/20}$

```python
def g(t):
    return np.exp(-0.05*t)
```

```python
fig = plt.figure(figsize=(8,5))
plt.title(r'$g(t)=e^{-t/20}$',y=1.01,size=20)
for alpha in np.linspace(0.1,0.9,9):
    T=[]
    Df=[]
    t=0.01
    while t<=10:
        T.append(t)
        Df.append(nfr.fd_RiemannLiouville(g,t,0,alpha,1e-6))
        t+=0.1
    plt.semilogy(T,Df,label=r'$\alpha='+str(round(alpha,1))+'$',linewidth=0.5+alpha*2)
plt.legend(loc=(1,0.2),shadow=False,ncol=1,prop={'size':16},frameon=False)
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel(r'$t$',size=23)
plt.ylabel(r'${}_{RL} D_{0,t}^\alpha\, g(t)$',size=23)
plt.show()
fig.savefig('RiemannLiouville1.png',dpi=100,bbox_inches='tight')
```

### Example: Fractional Derivative of $g(t)=\cos(t)$

```python
def g(t):
    return t*np.cos(t)
```

```python
fig = plt.figure(figsize=(8,5))
plt.title(r'$g(t)=t\cos(t)$',y=1.01,size=20)
for alpha in np.linspace(0.1,0.9,9):
    T=[]
    Df=[]
    t=0.001
    while t<10:
        T.append(t)
        Df.append(nfr.fd_RiemannLiouville(g,t,0,alpha,1e-6))
        t+=0.01
    plt.plot(T,Df,label=r'$\alpha='+str(round(alpha,1))+'$',linewidth=0.5+alpha*2)
plt.legend(loc=(1,0.2),shadow=False,ncol=1,prop={'size':16},frameon=False)
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel(r'$t$',size=23)
plt.ylabel(r'${}_{RL} D_{0,t}^\alpha\, g(t)$',size=23)
plt.show()
fig.savefig('RiemannLiouville2.png',dpi=100,bbox_inches='tight')
```
